```python
import numpy as np
p=np.arange(0.0,0.5,0.1)
s=np.exp(-p)*np.sin(2*np.pi*p)
plt.subplot(2,2,3)
plt.plot(p,s,color='k',label='graph p vs s')
plt.xlabel('p')
plt.ylabel('s')
plt.title('graph with sin wave')
plt.legend()
plt.show()
```

```python
p=np.arange(0.0,0.5,0.1)
s=np.exp(-p)*np.sin(2*np.pi*p)
plt.subplot(2,2,1)
plt.plot(p,s,color='k',label='1st graph with sin wave')
plt.xlabel('p')
plt.ylabel('s')
plt.title('sin wave graph with null values')
plt.legend()
p[0]=np.nan
p[1]=np.nan
plt.subplot(2,1,2)
plt.show()
```
