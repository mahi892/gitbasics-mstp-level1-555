```python
from sklearn.datasets import load_boston
```

```python
boston=load_boston()
boston
```

```python
boston.data
```

```python
boston.target
```

```python
boston.feature_names
```

```python
#creating dataset
import pandas as pd
boston_df=pd.DataFrame(boston.data,columns=boston.feature_names)
```

```python
boston_df.head()
```

```python
boston_df["Land Rate"]=boston.target
boston_df.head()
```

```python
x=boston_df.drop("Land Rate",axis=1).values
x
```

```python
y=boston_df["Land Rate"]
y.values
```

```python
    #Ridge Regression=sum of error+sum of squres of magnitude of
    # coeffient(alpha)
from sklearn.linear_model import Ridge
rr=Ridge(alpha=1,solver='svd')
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=2)
```

```python
rr.fit(x_train,y_train)
```

```python
y_predict=rr.predict(x_train)
y_predict
```

```python
#score
rr.score(x_train,y_train)
```

```python
from sklearn.linear_model import LinearRegression
lin=LinearRegression()
```

```python
lin.fit(x_train,y_train)
```

```python
rr.coef_
```

```python
lintrain_score=lin.score(x_train,y_train)
print("lin Testing score", lintrain_score)
```

```python
lin.coef_
```

```python
import matplotlib.pyplot as plt
plt.plot(rr.coef_,marker="*",label="Ridge coefficient",color="blue")
plt.plot(lin.coef_,marker=".",label="Linear Coeffi",color="y")
plt.xlabel("coefficient index")
plt.ylabel("coefficient magnitude")
plt.title("Ridge Vs Linear")
plt.legend()
plt.show()
```
