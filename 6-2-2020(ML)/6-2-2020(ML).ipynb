```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
ad_df=pd.read_csv("Advertisement.csv")
ad_df
```

```python
ad_df.head()
```

```python
ad_df.drop('Unnamed: 0',axis=1,inplace=True)
ad_df.head()
```

```python
x=ad_df.iloc[:,0:3].values
y=ad_df.iloc[:,3:].values
print(x.shape)
print(y.shape)
```

```python
from sklearn.linear_model import LinearRegression
lin=LinearRegression()
```

```python
lin.fit(x,y)
```

```python
y_predict=lin.predict(x)
y_predict
```

```python
lin.predict([[180.8,10.8,58.4]])
```

```python
plt.scatter(ad_df["radio"].values,y,c='r',label="Before prediction")
plt.plot(x,y_predict,c='y',label="After prediction")
plt.xlabel("Cost of Tv,radio,news paper")
plt.ylabel("sales")
plt.title("LinearRegression")
plt.legend(loc="best")
plt.show()
```

```python
from sklearn.preprocessing import PolynomialFeatures
poly_feat=PolynomialFeatures()
```

```python
x_poly=poly_feat.fit_transform(x)
x_poly
```

```python
lin.fit(x_poly,y)
```

```python
y_prediction=lin.predict(x_poly)
y_prediction
```

```python
lin.intercept_
```

```python
lin.coef_
```

```python
plt.scatter(ad_df["TV"],y,c='r',label="Before prediction")
plt.plot(x,y_predict,c='y',label="After prediction with LinearRegression")
plt.xlabel("Cost of Tv,radio,news paper")
plt.ylabel("sales")
plt.title("LinearRegression")
plt.legend(loc="best")
plt.show()
```
