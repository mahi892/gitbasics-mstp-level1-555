```python
# draw a plot for time and distance by using list comphrension
# distance=2*time
import matplotlib.pyplot as h
time=[i for i in range(2,5)]
distance=[j*2 for j in time]
h.plot(time,distance,'ro')
h.xlabel('x-time')
h.ylabel('y-distance')
h.title('time&distance')
h.show()
```

#### matplotlib ticks



```python
##ticks is values within a coordinates
import matplotlib.pyplot as pl
x=[5,10,15,20,25,30,35,40,45,50]
print(x)
y=[9,8,7,6,5,4,3,2,1,10]
print(y)
pl.plot(y,x)
pl.title('A Graph a withexample to ticks')
pl.xlabel('natural Numbers')
pl.ylabel('values with ++5 up to 50')
```

```python
import matplotlib.pyplot as pl
import numpy as np
pl.xticks(np.arange(1,11))
pl.yticks(np.arange(5,55,5))
pl.plot(x,y)
pl.title(" maplotlib-ticks")
pl.xlabel("Natural Numbers")
pl.ylabel('5-50 values with 5+')
pl.show
```

## representing a sine wave
A=sine(wt)
A=Amplitude
w=frequency
t=time

```python
import numpy as np
import matplotlib.pyplot as pl
t=np.arange(1,10,0.1)
Amplitude=np.sin(t)
pl.grid()
pl.plot(t,Amplitude)
pl.title("sine wave A=sin(wt)")
pl.xlabel('t')
pl.ylabel('Amplitude')
pl.show()
```

```python
import pandas as pd
df=pd.read_csv('income.csv')
df
```

```python
import matplotlib.pyplot as pl
years=df.columns[2::]
Alaska_income=df.iloc[1,2:]
pl.plot(years,Alaska_income)
pl.title("Income of Alaska from 2005-2013")
pl.xlabel("years")
pl.ylabel("Income of Alaska")
pl.savefig("Income of Alaska from 2005-2013")
```

```python
# plot for all status income from 2005-2013
import matplotlib.pyplot as pl
import pandas as pd
years=df.columns[-5::]
allstates_income=df.iloc[0:,2:]
allstates_income
pl.plot(years,allstates_income)
pl.xlabel("years")
pl.ylabel("states")
pl.title("allstate_income")
pl.show()
```
