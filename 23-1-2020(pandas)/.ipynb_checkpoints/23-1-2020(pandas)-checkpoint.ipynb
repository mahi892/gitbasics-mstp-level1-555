```python
help(plt.plot)
```

```python
x=[1,2,3,4]
y=[5,6,7,3]
plt.plot(x,y)
```

```python
plt.plot(x,y,linewidth=1.5,markersize=10,color='green')
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("graphing")
```

```python
plt.plot(x,y,'s',linewidth=1.5,markersize=10,color='green')
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("graphing")
```

```python
x1=[6,7,8,3]
y1=[4,7,9,2]
plt.bar(x1,y1)
plt.bar(x,y,color="red")
```

```python
plt.bar([1,2,3,4],[5,6,7,8],color='m')
plt.bar([2,4,3,1],[7,5,8,6],color='r')
```

```python
help(plt.hist)
```

```python
bins=[1,2,4,6,8,9]
plt.hist(x,bins,histtype='step',orientation='horizontal')
```

```python
bins=[1,2,4,6,8,9]
plt.hist(x,bins,histtype='bar',orientation='horizontal')
```

```python
bins=[1,2,4,6,8,9]
plt.hist(x,bins,histtype='bar',orientation='horizontal',rwidth=2,color='m')
```
