```python
import seaborn as sns
```

```python
tips=sns.load_dataset("tips")
tips.head()
```

```python
help(sns.factorplot)
```

```python
sns.factorplot(x="total_bill",y="day",data=tips)
```

```python
sns.factorplot(x="total_bill",y="day",hue="sex",markers=['x','s'],data=tips,dodge=True)
```

```python
from numpy import median
sns.factorplot(x="total_bill",y="day",hue="sex",markers=['x','s'],data=tips,dodge=True,estimator=median)
```

```python
sns.factorplot("sex","tip",hue="day",data=tips)
```

```python
sns.factorplot("sex","tip",hue="day",data=tips,markers=['x','s','D','o'])
```

```python
help(sns.pointplot)
```

```python
sns.pointplot("total_bill","sex",data=tips,hue='day',markers=['x','s','D','o'])
```

```python
sns.pointplot(x="time",y="total_bill",data=tips,hue='sex',markers=['x','s'],
              linestyles=["-","-."])
```

```python
sns.pointplot(x="tip",y='day',data=tips,hue="sex",join=False,dodge=True)
```

```python
sns.pointplot(x="tip",y='day',data=tips,hue="sex",join=False,dodge=True,estimator=median)
```

```python
#bar plot
t=sns.load_dataset('titanic')
t.tail(10)
```

```python
sns.barplot(x="class",y="survived",data=t,hue='sex',palette="RdBu")
```

```python
##colors in seaborn
import seaborn as sns
sns.color_palette()
```

```python
sns.palplot(sns.color_palette())
```

```python
c=sns.color_palette("RdBu")
sns.palplot(c)
```

```python
c=sns.color_palette("RdBu",n_colors=10)
sns.palplot(c)
```

```python
c=sns.color_palette("Blues",n_colors=10)
sns.palplot(c)
```

```python
c=sns.color_palette(sns.color_palette("ch:3"))
sns.palplot(c)
```

```python
c=sns.color_palette(sns.color_palette("ch:5,r=-1,l=-1"))
sns.palplot(c)
```

```python
sns.set()
cc=sns.cubehelix_palette()
cc
```

```python
sns.palplot(cc)
```

```python
sns.palplot(sns.cubehelix_palette(reverse=True))
```

```python
sns.palplot(sns.cubehelix_palette(start=2,rot=-1,reverse=True))
```

```python
import matplotlib.pyplot as plt
sns.heatmap(sns.cubehelix_palette())
plt.show()
```

```python
import numpy as np
x=np.arange(16).reshape(4,4)
cc=sns.cubehelix_palette()
sns.heatmap(x,cmap=cc)
```

```python
x=np.arange(16).reshape(4,4)
cc=sns.cubehelix_palette(dark=-2,light=1)
sns.heatmap(x,cmap=cc)
```
